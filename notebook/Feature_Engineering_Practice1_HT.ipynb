```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import time
```

```python
df = pd.read_csv('../datasets/train.csv')
```

```python
df = pd.get_dummies(df)
df_train = df[df['locdt']<72]
df_test = df[df['locdt']>=72]
print(f'df_train:{df_train.shape}')
print(f'df_test:{df_test.shape}')
assert df_train.shape[0]+df_test.shape[0] == 1521787
```

```python
df_train = df_train.sort_values(by=['bacno','cano','locdt','loctm'],ascending=True)
```

## 新增一欄 num_trades_3d 紀錄該卡號(cano發生這筆交易的前三天，有多少次交易

```python
init_time = time.time()
df_train['num_trades_3d'] = 0
for i in range(100000): # range(10000)花了176秒 range(100000)花了1900秒 這樣算起來全部資料1200000要花6-7小時以上 太久
    dt = df_train.iloc[i,10]      # 定位點
    for ii in range(i-1,i-1000,-1): # 回頭看
        if (df_train.iloc[ii,10] >= dt-3) & (df_train.iloc[ii,2]==df_train.iloc[i,2]):  # 日期符合&卡號一樣
            df_train.iloc[i,28] = df_train.iloc[i,28]+1
        elif (df_train.iloc[ii,10] < dt-3) or (df_train.iloc[ii,2]!=df_train.iloc[i,2]):
            break  
print(f'Elapsed time:{time.time()-init_time} sec.')
```
